```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
reliance = yf.Ticker("RELIANCE.NS")
data = reliance.history(period="5y")  # Fetch 5 years of data
```

```python
data['Daily Return'] = data['Close'].pct_change()
```

```python
plt.figure(figsize=(10, 6))
data['Daily Return'].plot(title="Daily Returns of Reliance Industries")
plt.show()
```

```python
fig = go.Figure(data=[go.Candlestick(
        x=data.index,
        open=data['Open'],
        high=data['High'],
        low=data['Low'],
        close=data['Close']
    )])
fig.update_layout(title="Candlestick Chart of Reliance Industries")
fig.show()
```

```python
import streamlit as st
import yfinance as yf

st.title("Indian Stock Market Analysis")

ticker = st.text_input("Enter Stock Ticker (e.g., RELIANCE.NS):", "RELIANCE.NS")
data = yf.Ticker(ticker).history(period="1y")

st.line_chart(data['Close'])
st.write("Data Summary:", data.describe())
```
